```python
# Dropping (BPXOPLS1, BPXOPLS2, BPXOPLS3) as these are not crucial for hypertension prediction
df1.drop(columns=['BPXOPLS1', 'BPXOPLS2', 'BPXOPLS3'], inplace=True)
df1
```

```python
# Removing the  NaNs:
df1.dropna(subset=['BPXOSY1', 'BPXODI1'], inplace=True)  # Drop rows missing key BP readings
df1
```

```python
# Define the BP columns
bp_columns = ['BPXOSY1', 'BPXODI1', 'BPXOSY2', 'BPXODI2', 'BPXOSY3', 'BPXODI3']

# Fill missing values with row-wise mean
df1[bp_columns] = df1[bp_columns].apply(lambda row: row.fillna(row.mean()), axis=1)
```

```python
df1.isna().sum() # Final check of the missing value
```

```python
df2.isna().sum()
```

```python
# Columns with too many missing values (more than 50%)
cols_to_drop = [
    'BMIWT', 'BMIRECUM', 'BMXHEAD', 'BMIHEAD', 'BMIHT', 
    'BMILEG', 'BMIARML', 'BMIARMC', 'BMIWAIST', 'BMIHIP'
]

df2 = df2.drop(columns=cols_to_drop)
```

```python
# Filling missing values for key features
# 1. Numerical features

# Using the median for numerical columns
cols_to_fill_median = ['BMXWT', 'BMXHT', 'BMXBMI', 'BMXWAIST', 'BMXHIP']
df2[cols_to_fill_median] = df2[cols_to_fill_median].fillna(df2[cols_to_fill_median].median())


```

```python
# Ussing the mode (most frequent value) for categorical columns
df2['BMDBMIC'] = df2['BMDBMIC'].fillna(df2['BMDBMIC'].mode()[0])
```

```python
df2.isna().sum()  # Check for remaining missing values
```

```python
# Dropping the columns that has high number of missing values
df2 = df2.drop(columns=['BMXRECUM'])
```

```python
# Filling missing values for important features
# Since BMXLEG, BMXARML, and BMXARMC are continuous numerical values, we fill them using the median (less affected by outliers).

cols_to_fill_median = ['BMXLEG', 'BMXARML', 'BMXARMC']
df2[cols_to_fill_median] = df2[cols_to_fill_median].fillna(df2[cols_to_fill_median].median())
```

```python
# Final check of the missing values in 2nd dataset
df2.isna().sum()
```

### Merging the data: 
##### Sincee both dataset contains the column SEQN (Respondant Sequence Number) as a unique identifier, we can merge them. We'll be joining the data with inner join as it will ensure we only keep respondents who have both blood pressure and BMI measurements and avoids issues with missing data.


```python
# Merge on SEQN
df_merged = pd.merge(df1, df2, on="SEQN", how="inner")  # Using "inner" to keep only matching records

# Check shape of merged data
print(f"Merged Dataset Shape: {df_merged.shape}")

# Display first few rows
df_merged.head()
```

```python
# Checking the missing value in the mereged data
df_merged.isna().sum()
```

```python
# Saving the data to csv file
df_merged.to_csv('merged_data.csv', index=False)
```
