```python
# Libraries 
import pandas as pd
import numpy as np
```

```python
# Loading the dataset
blood_pressure_df = pd.read_csv(r'C:\Users\jugal\OneDrive\Documents\Ironhack_project\Final_project\merged_data.csv')
demo_df =  pd.read_csv(r'C:\Users\jugal\OneDrive\Documents\Ironhack_project\Final_project\demo_data.csv')
```

```python
# Checking the common columns that can be used for merging
print("Blood Pressure Dataset Columns:", blood_pressure_df.columns.tolist())
print("\nDemographic Dataset Columns:", demo_df.columns.tolist())
```
