### Renming the column name

```python
new_column_names = {
    'BPAOARM': 'ArmCircumference',
    'BPAOCSZ': 'CuffSize',
    'BPXOSY1': 'SystolicBP1',
    'BPXODI1': 'DiastolicBP1',
    'BPXOSY2': 'SystolicBP2',
    'BPXODI2': 'DiastolicBP2',
    'BPXOSY3': 'SystolicBP3',
    'BPXODI3': 'DiastolicBP3',
    'BMDSTATS': 'BodyMeasurementStatus',
    'BMXWT': 'Weight',
    'BMXHT': 'Height',
    'BMXBMI': 'BMI',
    'BMDBMIC': 'BMICategory',
    'BMXLEG': 'LegLength',
    'BMXARML': 'ArmLength',
    'BMXARMC': 'ArmCircumference',
    'BMXWAIST': 'WaistCircumference',
    'BMXHIP': 'HipCircumference'
}

df = df.rename(columns=new_column_names)
```

```python
df['ArmCircumference'] = df['ArmCircumference'].apply(pd.to_numeric, errors='coerce') # Converts to number, sets invalid to NaN

# Prepare the mapping dictionaries
gender_map = {'Male': 1, 'Female': 0}
ethnicity_map = {
    'Mexican American': 1, 
    'Other Hispanic': 2,
    'Non-Hispanic White': 3, 
    'Non-Hispanic Black': 4, 
    'Other Race': 5
}

# Applying the transformation
df = df.assign(
    Gender = df['Gender'].map(gender_map),
    Ethnicity = df['Ethnicity'].map(ethnicity_map)
)
```

```python
# Drop columns with more than 30% missing values
df = df.dropna(thresh=len(df) * 0.7, axis=1)
df
```
