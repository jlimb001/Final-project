### Handling the overpopulating issue

```python
def preprocess_data(data_path):
    '''Preprocess NHANES data for hypertension prediction'''
    df = pd.read_csv(data_path)

    # Create target variable (hypertension)
    df['avg_systolic'] = df[['SystolicBP1', 'SystolicBP2', 'SystolicBP3']].mean(axis=1)
    df['avg_diastolic'] = df[['DiastolicBP1', 'DiastolicBP2', 'DiastolicBP3']].mean(axis=1)

    # Define target
    df['Hypertension'] = ((df['avg_systolic'] >= 130) | (df['avg_diastolic'] >= 80)).astype(int)
    target = 'Hypertension'

    # Define features
    numerical_features = ['Weight', 'Height', 'BMI', 'WaistCircumference', 'HipCircumference',
                          'ArmCircumference', 'Age']
    categorical_features = ['Gender', 'Ethnicity', 'Education']

    # Handles missing values
    df[numerical_features] = SimpleImputer(strategy='mean').fit_transform(df[numerical_features])
    df[categorical_features] = SimpleImputer(strategy='most_frequent').fit_transform(df[categorical_features])

    # Feature Engineering
    df['waist_hip_ratio'] = df['WaistCircumference'] / df['HipCircumference']
    df['bmi_age_factor'] = df['BMI'] * df['Age']

    # Final feature set
    numerical_features += ['waist_hip_ratio', 'bmi_age_factor']
    X = df[numerical_features + categorical_features]
    y = df[target]

    # Handle imbalance
    if y.value_counts(normalize=True).min() < 0.3:
        print("Applying SMOTE for imbalance handling...")
        smote = SMOTE(random_state=42)
        X, y = smote.fit_resample(X, y)

    return X, y, numerical_features, categorical_features
```

```python
def build_preprocessor(numerical_features, categorical_features):
    '''Preprocessing pipeline'''
    num_pipeline = Pipeline([
        ('imputer', SimpleImputer(strategy='mean')),
        ('scaler', StandardScaler())
    ])

    cat_pipeline = Pipeline([
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
    ])

    preprocessor = ColumnTransformer([
        ('num', num_pipeline, numerical_features),
        ('cat', cat_pipeline, categorical_features)
    ])

    return preprocessor
```

```python
def train_with_early_stopping(X_train, X_val, y_train, y_val):
    '''Train XGBoost with early stopping'''
    # Pass early_stopping_rounds to the constructor
    model = XGBClassifier(
        n_estimators=1000,
        learning_rate=0.05,
        max_depth=5,
        eval_metric='auc',
        use_label_encoder=False,
        random_state=42,
        early_stopping_rounds=10  # Moved here
    )

    # Fit the model with validation set
    model.fit(
        X_train,
        y_train,
        eval_set=[(X_val, y_val)],  # Validation set for evaluation
        verbose=True
    )

    return model
```

```python
def evaluate_model(model, X_test, y_test):
    '''Evaluate model and plot learning curve'''
    y_prob = model.predict_proba(X_test)[:, 1]
    y_pred = model.predict(X_test)
    roc_auc = roc_auc_score(y_test, y_prob)
    pr_auc = average_precision_score(y_test, y_prob)

    print(f"ROC-AUC: {roc_auc:.4f}")
    print(f"PR-AUC: {pr_auc:.4f}")
    # Compute confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    
    # Display confusion matrix
    print("\nConfusion Matrix:")
    print(cm)
    
    # Plot confusion matrix as heatmap
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=["No Hypertension", "Hypertension"], 
                yticklabels=["No Hypertension", "Hypertension"])
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.title("Confusion Matrix")
    plt.show()
    
    # Additional metrics from confusion matrix
    tn, fp, fn, tp = cm.ravel()
    accuracy = (tp + tn) / (tp + tn + fp + fn)
    precision = tp / (tp + fp) if (tp + fp) > 0 else 0
    recall = tp / (tp + fn) if (tp + fn) > 0 else 0
    f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
    
    # Print additional metrics
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1-Score: {f1_score:.4f}")
```

```python
def compare_models(X_train, X_val, y_train, y_val):
    '''Compare multiple models'''
    # Define models
    models = {
        "Logistic Regression": LogisticRegression(random_state=42),
        "Decision Tree": DecisionTreeClassifier(random_state=42),
        "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
        "XGBoost": XGBClassifier(n_estimators=100, learning_rate=0.05, max_depth=5,
                                 eval_metric='auc', use_label_encoder=False, random_state=42),
        'SVM': SVC(probability=True, random_state=42)
    }

    # Train and evaluate each model
    results = []
    for name, model in models.items():
        print(f"Training {name}...")
        model.fit(X_train, y_train)
        
        # Predict probabilities for validation set
        y_prob = model.predict_proba(X_val)[:, 1]
        y_pred = model.predict(X_val)
        
        # Calculate metrics
        roc_auc = roc_auc_score(y_val, y_prob)
        pr_auc = average_precision_score(y_val, y_prob)
        results.append({"Model": name, "ROC-AUC": roc_auc, "PR-AUC": pr_auc})
        
        print(f"{name} ROC-AUC: {roc_auc:.4f}, PR-AUC: {pr_auc:.4f}")
        # Compute confusion matrix for validation set
        cm = confusion_matrix(y_val, y_pred)
        
        # Display confusion matrix as heatmap
        print("\nConfusion Matrix (Validation Set):")
        print(cm)
        
        plt.figure(figsize=(6, 4))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=["No Hypertension", "Hypertension"], 
                    yticklabels=["No Hypertension", "Hypertension"])
        plt.xlabel("Predicted")
        plt.ylabel("Actual")
        plt.title(f"Confusion Matrix: {name}")
        plt.show()
        
        # Additional metrics from confusion matrix
        tn, fp, fn, tp = cm.ravel()
        accuracy = (tp + tn) / (tp + tn + fp + fn)
        precision = tp / (tp + fp) if (tp + fp) > 0 else 0
        recall = tp / (tp + fn) if (tp + fn) > 0 else 0
        f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
        
        print(f"Accuracy: {accuracy:.4f}")
        print(f"Precision: {precision:.4f}")
        print(f"Recall: {recall:.4f}")
        print(f"F1-Score: {f1_score:.4f}")

    # Convert results to DataFrame for comparison
    results_df = pd.DataFrame(results)
    return results_df
```

```python
# Path to your dataset 
data_path = r'C:\Users\jugal\OneDrive\Documents\Ironhack_project\Final_project\feature_engineering.csv'

# Load and preprocess data
X, y, num_feats, cat_feats = preprocess_data(data_path)

# Build preprocessing pipeline
preprocessor = build_preprocessor(num_feats, cat_feats)

# Transform features using the preprocessor pipeline
X_transformed = preprocessor.fit_transform(X)

# Train/Validation/Test split (60% train, 20% validation, 20% test)
X_temp, X_test, y_temp, y_test = train_test_split(X_transformed, y, test_size=0.2, stratify=y, random_state=42)

# Split training and validation sets (75% of temp data for training and 25% for validation)
X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size=0.25, stratify=y_temp)

# Compare models
results_df = compare_models(X_train, X_val, y_train, y_val)

print("\nModel Comparison Results:")
print(results_df)
```

```python
# Define parameter grid
param_grid = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.05, 0.1],
    'max_depth': [3, 4, 5],  
    'min_child_weight': [3, 5, 7], 
    'subsample': [0.6, 0.8, 1],
    'colsample_bytree': [0.6, 0.8, 1]
}
```

```python
# Initialize model
xgb = XGBClassifier(eval_metric='auc', use_label_encoder=False)
```

```python
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

random_search = RandomizedSearchCV(
    estimator=xgb,
    param_distributions=param_grid,
    scoring='roc_auc',
    n_iter=50,
    cv=cv,
    verbose=2,
    random_state=42,
    n_jobs=-1
)
```

```python
# Fit on training data
random_search.fit(X_train, y_train)

# Best parameters
print("Best Parameters:", random_search.best_params_)

# Evaluate on validation set
best_xgb = random_search.best_estimator_
evaluate_model(best_xgb, X_test, y_test)
```

```python
# from sklearn.svm import SVC
# from sklearn.model_selection import GridSearchCV

# def tune_svm(X_train, y_train):
#     '''Hyperparameter tuning for SVM'''
    
#     # Define parameter grid
#     param_grid = {
#         'C': [0.1, 1, 10, 100],  # Regularization strength
#         'kernel': ['linear', 'rbf', 'poly'],  # Kernel types
#         'gamma': ['scale', 'auto'],  # Kernel coefficient
#         'degree': [2, 3, 4]  # Only used for 'poly' kernel
#     }

#     # Initialize SVM model
#     svm = SVC(probability=True, random_state=42)

#     # Perform Grid Search with PR-AUC scoring
#     grid_search = GridSearchCV(
#         estimator=svm,
#         param_grid=param_grid,
#         scoring='average_precision',  # Optimize for PR-AUC
#         cv=5,  # 5-fold cross-validation
#         verbose=2,
#         n_jobs=-1
#     )

#     # Fit on training data
#     grid_search.fit(X_train, y_train)

#     print("Best Parameters:", grid_search.best_params_)
    
#     return grid_search.best_estimator_

from sklearn.svm import SVC
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import uniform, randint

def tune_svm(X_train, y_train):
    '''Hyperparameter tuning for SVM using RandomizedSearchCV'''

    # Define parameter distributions
    param_distributions = {
        'C': [0.1, 1, 10],
        'kernel': ['linear', 'rbf'],
        'gamma': ['scale', 'auto']
    }

    # Initialize SVM model
    svm = SVC(probability=True, random_state=42)

    # Perform Randomized Search (faster than Grid Search)
    random_search = RandomizedSearchCV(
        estimator=svm,
        param_distributions=param_distributions,
        n_iter=6,  # Number of different combinations to try
        scoring='average_precision',
        cv=3,  # 3-fold cross-validation to speed it up
        n_jobs=-1,
        verbose=2,
        random_state=42
    )

    # Fit on training data
    random_search.fit(X_train, y_train)

    print("Best Parameters:", random_search.best_params_)
    return random_search.best_estimator_

```

```python
def train_and_evaluate_svm(X_train, X_val, y_train, y_val, X_test, y_test):
    '''Train and evaluate the best SVM model'''
    
    # Tune hyperparameters
    best_svm = tune_svm(X_train, y_train)
    
    # Evaluate on validation set
    print("\nEvaluating on Validation Set:")
    evaluate_model(best_svm, X_val, y_val)
    
    # Evaluate on test set
    print("\nEvaluating on Test Set:")
    evaluate_model(best_svm, X_test, y_test)
    
    return best_svm
```
