```python
# Load data

#Demographic dataset
demographic_df = pd.read_sas(r'C:\Users\jugal\OneDrive\Documents\Ironhack_project\Final_project\P_DEMO.xpt')

# Cleaned dataset of blood pressure
df = pd.read_csv(r'C:\Users\jugal\OneDrive\Documents\Ironhack_project\Final_project\merged_data.csv')

```

```python
# Changing the name to deal with the missing values
demographic_df.rename(columns={'RIDAGEYR': 'Age', 'DMDEDUC2': 'Education', 'INDFMPIR': 'incom_to_pov_ratio', 'SDDSRVYR': 'Survery_year'}, inplace = True)
```

```python
# changing the name to gender and converting it to categorical value
demographic_df['Gender'] = demographic_df['RIAGENDR'].map({1: 'Male', 2: 'Female'})
```

```python
# Filling the missing value on education column with placeholder
demographic_df.fillna({'Education':'Unknowm'}, inplace=True)
```

```python
# Fillin the missing value of income-to-poverty ratio column with median
demographic_df['incom_to_pov_ratio'].fillna(demographic_df['incom_to_pov_ratio'].median(), inplace=True)
```

```python
# Ethnicit column: converting it to readable categories
ethnicity_map = {
    1: 'Mexican American',
    2: 'Other Hispanic',
    3: 'Non-Hispanic White',
    4: 'Non-Hispanic Black',
    5: 'Other Race'
}
demographic_df['Ethnicity'] = demographic_df['RIDRETH1'].map(ethnicity_map)
```

```python
# Filtering the data
demographic_df = demographic_df[['SEQN', 'Survery_year', 'Age', 'Gender', 'Education', 'Ethnicity','incom_to_pov_ratio']]
```
